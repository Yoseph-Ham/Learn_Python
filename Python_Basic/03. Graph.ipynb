### 기초 그래프 그리기 - 히스토그램, 산점도, 박스 그래프

```python
import seaborn as sns

tips = sns.load_dataset("tips")
print(tips.head())
print(type(tips))
# 지불금액, 팁, 성별, 흡연여부, 요일, 시간, 방문인원
```

```python
#fig(기본틀), 그래프 격자(axes1)
%matplotlib inline
import matplotlib.pyplot as plt

fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)
```

## 1. 히스토그램은                                                                   df의 열 데이터 분포와 빈도를 살펴보는 용도
### 변수를 하나만 사용해서 그린 '일변량 그래프'

```python
axes1.hist(tips['total_bill'], bins = 10) # bins = 10 x축의 간격을 10
axes1.set_title("Histogram of Total Bill")
axes1.set_xlabel("Frequency")
axes1.set_ylabel("Total Bill")

fig
```

## 2. 산점도 그래프
### 변수 2개 사용한 '이변량 그래프' - total_bill 열에 따른  tip 열의 분포를 나타낸 산점도 그래프

```python
#scatter 메서드에 total_bill, tips 열을 전달
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

## 3. 박스그래프
    *. 이산형 변수 - Female, Male처럼 명확하게 구분되는 값
    *. 연속형 변수 - Tip와 같이 명확하게 셀 수 없는 변위의 값

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1, 1, 1)

axes1.boxplot(
[tips[tips['sex'] == 'Female']['tip'],
 tips[tips['sex'] == 'Male']['tip']],
    labels=['Female', 'Male'])

axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

## 4. 다변량 데이터로 다변량 그래프 그리기 - 산점도 그래프
*. 다변량 - 3개 이상의 변수를 사용

```python
# 성별을 새 변수로 추가(문자열은 산점도 그래프와 색상을 지정하는 값으로 사용할 수 없다.)
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
```

```python
# apply 메서드 - 사용자가 작성한 함수를 df의 각 행과 열에 적용하는 메서드
tips['sex_color'] = tips['sex'].apply(recode_sex) # recode_sex의 값 사용하여 sex_color 칼럼 추가
```

```python
tips
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(
    x = tips['total_bill'],
    y = tips['tip'],
    s = tips['size'] * 15, # 점의 크기
    c = tips['sex_color'], # 점의 색
    alpha = 0.8) # 투명도

axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

## 단변량 그래프 그리기 - 히스토그램

### seaborn 라이브러리는 matplotlib 라이브러리를 기반으로 만든 라이브러리

```python
import seaborn as sns
tips = sns.load_dataset("tips")
```

## seaborn 라이브러리로 히스토그램
*seaborn 라이브러리로 히스토그램을 그리려면  subplots, distplot 메서드를 사용

*distplot 메서드 : 기본 틀 만듬 / total_bill 메서드 : total_bill 열데이터를 전달

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill histogram with Denesity Plot')
```

```python
# 밀집도 (kde - 정규화시켜 넓이가 1이 되도록 그린 그래프) 제외
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde = False)
ax.set_title('Total Bill histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
# 밀도 그래프만 남기기 (히스토그램 제외)
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist = False)
ax.set_title('Total Bill Denesity')
ax.set_xlabel('Total Bill')
ax.set_ylabel('')
```

```python
# rug 인자 추가 - 그래프의 축에 동일한 길이의 직선을 붙여
# 데이터의 밀집도를 표현
hist_den_rug, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug = True)
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total BIll')
```

## 다양한 종류의 이변량 그래프 그리기

### 1. seaborn 라이브러리로 산점도 그래프 그리기

*. regplot 메서드를 사용하면 산점도 그래프와 회귀선을 그릴 수 있다.

```python
ax = plt.subplots()
ax = sns.regplot(x = 'total_bill', y = 'tip', data = tips)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
# 회귀선 지우기 fit_reg = False

ax = plt.subplots()
ax = sns.regplot(x = 'total_bill', y = 'tip', data = tips, fit_reg = False)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
joint = sns.jointplot(x = 'total_bill', y = 'tip', data = tips)

joint.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')
joint.fig.suptitle('Joint plot of Total Bill and Tip', fontsize =10, y = 1.03)
```

```python
# hexbin : 육각 그래프 - 빈도가 높을 수록 색이 진해짐 kind = "hex"
hexbin = sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = "hex")

hexbin.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')
hexbin.fig.suptitle('Hexbin Joint plot of Total Bill and Tip', 
                    fontsize =10, y = 1.03)
```

## 2차원 밀집도 그래프 그리기

```python
ax = plt.subplots()
ax = sns.kdeplot(data = tips['total_bill'],
                data2 = tips['tip'],
                shade = True) # 음영효과

ax.set_title('Kernel Density plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

## 그래프에 스타일 적용하기

```python
sns.set_style('whitegrid')
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', hue = 'sex', data = tips, split = True)
```

```python
fig = plt.figure()
seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']
for idx, style in enumerate(seaborn_styles):
    #print(style)
    plot_position = idx + 1
    #print(plot_position)
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(x= 'time', y='total_bill', data = tips, ax = ax)
        violin.set_title(style)

fig.tight_layout()
```

## 파이 그래프 만들기 

```python
import matplotlib.pyplot as plt
%matplotlib inline

labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10] # 점유율
explode = (0, 0, 0.1, 0) # 간격
colors = ['hotpink', 'green', 'red', 'coral']

fig, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct = '%1.1f%%', 
       shadow=True, startangle=90, colors=colors)
ax1.axis('equal')
plt.show()

# shadow = 그림자
# autopct 소수점 첫째까지 %로 표현
```
