```python
pd.read_excel("../data/mental.xls", header=1, usecols='c:k') 
# usecols='C:K' - C:K 까지의 칼럼을 사용
```

```python
col_names = ['스트레스', '스트레스남학생', '스트레스여학생', '우울감경험률',
            '우울남학생', '우울여학생', '자살생각율', '자살남학생', '자살여학생']

pd.read_excel("../data/mental.xls", header=1, usecols="C:K", names=col_names)

# names = col_names - names는 col_names 값을 사용
```

```python
raw_data = pd.read_excel("../data/mental.xls", header=1, 
              usecols="C:K", names=col_names)

raw_data
```

```python
# (100- ra_data.loc[0]의 값으로 raw_data.loc[1] 생성
raw_data.loc[1] = 100. - raw_data.loc[0]
raw_data
```

```python
raw_data['응답'] = ['그렇다', '아니다'] # 응답 컬럼생성

raw_data
```

```python
# set_index : 인덱스를 다른 칼럼으로 교체
# inplace : raw_data의 내용 갱신 결정 (False로 하면 저장이 되지 않음)
# drop : 인덱스를 해체하고 남은 데이터를 사용 할 것인지 (False로 하면 데이터가 남음)

raw_data.set_index('응답', drop=True, inplace=True)

raw_data
```

### 최종 데이터

```python
raw_data = pd.read_excel("../data/mental.xls", header=1, 
              usecols="C:K", names=col_names)

raw_data.loc[1] = 100. - raw_data.loc[0]

raw_data['응답'] = ['그렇다', '아니다'] # 응답 컬럼생성

raw_data.set_index('응답', drop=True, inplace=True)

raw_data
```

# matplotlib의 한글 문제

```python
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False # 마이너스(-) 기호 깨지는 현상

# f_path = "/Library/Fonts/AppleGothic.ttf" # 맥
f_path = 'C:/Windows/Fonts/malgun.ttf' # 윈도우즈
font_name = font_manager.FontProperties(fname=f_path).get_name()
rc('font', family = font_name)
```

```python
raw_data['스트레스'].plot.pie()
plt.show()
```

```python
raw_data['스트레스'].plot.pie(explode=[0,0.02]);
#explode = 간격
```

```python
# matplotlib.pyplot.subplots는 figure와 axes로 반환
# f, ax로 받으면 figure와 axes의 속성을 각각 지정 가능
# subplot 3개이므로 ax는 ax[0], ax[1], ax[2] 속성 지정

f, ax = plt.subplots(1,3, figsize=(16,8))

raw_data['스트레스'].plot.pie(explode=[0,0.02], ax=ax[0], autopct='%1.1f%%')
ax[0].set_title('스트레스를 받은적이 있다')
ax[0].set_ylabel('')

raw_data['우울감경험률'].plot.pie(explode=[0,0.02], ax=ax[1], autopct='%1.1f%%')
ax[0].set_title('우울증을 경험한 적이 있다')
ax[0].set_ylabel('')

raw_data['자살생각율'].plot.pie(explode=[0,0.02], ax=ax[2], autopct='%1.1f%%')
ax[0].set_title('자살을 고민한 적이 있다')
ax[0].set_ylabel('')

plt.show()
```
