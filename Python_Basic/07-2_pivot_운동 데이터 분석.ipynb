## 운동을 하지 않는 이유 데이터 분석

```python
not_exercise = pd.read_excel('../../data/not_exercise.xls')
not_exercise.head()
```

```python
# 기간 삭제
not_exercise.drop(columns='기간', inplace=True)
not_exercise
```

```python
# 목적과 상관없는 인덱스 22부터 삭제
not_exercise.drop(index = range(22,53), inplace = True)
not_exercise
```

```python
not_exercise['대분류'] == '성별'
```

```python
# 대분류에서 '성별'인 행만 선택
not_ex_sex = not_exercise[not_exercise['대분류'] == '성별'].copy()
not_ex_sex
```

```python
not_ex_sex.drop(columns='대분류', inplace = True)
not_ex_sex
```

```python
# reset_index : 기존의 행 인덱스를 제거하고 인덱스를 데이터 열로 추가
# set_index : 기존의 행 인덱스를 제거하고 데이터 열 중 하나를 인덱스로 설정
not_ex_sex.set_index('분류', inplace = True)
not_ex_sex
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rc('font',family='Gulim') # 한글

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False # 마이너스(-) 기호 깨지는 현상

#f_path = "/Library/Fonts/AppleGothic.ttf" #맥 경로
f_path = 'C:/Windows/Fonts/malgun.ttf' #윈도우즈 경로
font_name = font_manager.FontProperties(fname=f_path).get_name()
rc('font', family = font_name)
```

```python
f, ax = plt.subplots(1,2, figsize = (16,8))

not_ex_sex['운동을 할 충분한 시간이 없어서'].plot.pie(
    explode=[0,0.02], ax=ax[0], autopct = '%1.1f%%')

ax[0]. set_title('운동을 할 충분한 시간이 없어서')
ax[0].set_ylabel(' ')

not_ex_sex['함께 운동을 할 사람이 없어서'].plot.pie(
    explode=[0,0.02], ax=ax[1], autopct = '%1.1f%%')

ax[1]. set_title('함께 운동을 할 사람이 없어서')
ax[1].set_ylabel(' ')

plt.show()
```

```python
not_ex_age = not_exercise[not_exercise['대분류'] == '연령별'].copy()
not_ex_age.drop(columns = '대분류', inplace = True)
not_ex_age.set_index('분류', inplace = True)
not_ex_age
```

```python
f, ax = plt.subplots(1,3, figsize = (16,8))
explode = [0.02, 0.02, 0.02, 0.02, 0.02, 0.02]

not_ex_age['운동을 할 충분한 시간이 없어서'].plot.pie(
    explode=explode, ax=ax[0], autopct = '%1.1f%%')

ax[0].set_title('운동을 할 충분한 시간이 없어서')
ax[0].set_ylabel(' ')

not_ex_age['함께 운동을 할 사람이 없어서'].plot.pie(
    explode=explode, ax=ax[1], autopct = '%1.1f%%')

ax[1].set_title('함께 운동을 할 사람이 없어서')
ax[1].set_ylabel(' ')

not_ex_age['운동을 할 만한 장소가 없어서'].plot.pie(
    explode=explode, ax=ax[2], autopct = '%1.1f%%')

ax[2].set_title('운동을 할 만한 장소가 없어서')
ax[2].set_ylabel(' ')

```
