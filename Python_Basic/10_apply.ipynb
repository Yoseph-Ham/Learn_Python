# apply 메서드 활용
* apply 메서드는 사용자가 작성한 함수를 한 번에 데이터 프레임의 각 행과 열에 적용하여 실행할 수 있게 해주는 메서드

## 제곱 함수와 n 제곱 함수 만들기

```python
def my_sq(x):
    return x ** 2
```

```python
def my_exp(x, n):
    return x ** n
```

```python
print(my_sq(4))
```

```python
print(my_exp(3, 4))
```

## 시리즈와 데이터프레임에 apply 메서드 사용하기

### 시리즈와 apply 메서드

```python
import pandas as pd

df = pd.DataFrame({'a': [10,20,30], 'b':[20,30,40]})

print(df)
```

```python
# a열을 제곱해서 얻은 결과
print(df['a'] ** 2)
```

```python
# apply 메서드에 제곱 함수를 전달하여 시리즈의 모든 데이터에 제곱 함수 적용

sq = df['a'].apply(my_sq)
print(sq)
```

```python
# apply 메서드의 첫 번째 인자에 n 제곱 함수(my_exp)를 전달하여 n 제곱 함수를 적용
ex = df['a'].apply(my_exp, n=2)
print(ex)
```

```python
ex = df['a'].apply(my_exp, n=3)
print(ex)
```

## 데이터프레임의 누락값을 처리한 다음 apply 메서드 사용하기

### 1. 데이터프레임의 누락값 처리하기 - 열 방향

```python
import seaborn as sns
titanic = sns.load_dataset("titanic")
```

```python
print(titanic.info())
```

```python
import numpy as np

def count_missing(vec):
    null_vec = pd.isnull(vec) # 누락값을 확인
    null_count = np.sum(null_vec) # 누락값의 갯수를 셈
    return null_count
```

```python
cmis_col = titanic.apply(count_missing)
print(cmis_col)
```

```python
# count_missing 함수를 이용하여 누락값의 개수를 구하고
# size 속성을 이용하여 데이터프레임의 전체 데이터 수를 구하여 나누면 누락값의 비율을 계산

def prop_missing(vec):
    num = count_missing(vec)
    dem = vec.size
    #print('누락값의 갯수 (num)-> %d, 전체 데이터 수(dem) ->%d' %(num, dem))
    return num / dem
```

```python
pmis_col = titanic.apply(prop_missing)
print(pmis_col)
```

```python
# 누락값이 아닌 데이터 비율을 구함
# 전체(1) - 누락값의 비율

def prop_complete(vec):
    return (1 - prop_missing(vec))

print(titanic.apply(prop_complete))
```

## 8. 데이터프레임의 누락값을 처리하기 - 행 방향

```python
cmis_row = titanic.apply(count_missing, axis=1)
pmis_row = titanic.apply(prop_missing, axis=1)
pcom_row = titanic.apply(prop_complete, axis=1)

print(cmis_row.head())
```

```python
print(pmis_row.head())
```

```python
print(pcom_row.head())
```

```python
# 누락값의 개수를 구하여 데이터프레임에 'num_missing'열 추가
titanic['num_missing'] = titanic.apply(count_missing, axis = 1)

print(titanic.head())
```

```python
# 누락값이 있는 데이터만 모아옴, 누락값이 2개 이상인 데이터만 추출

print(titanic.loc[titanic.num_missing > 1, :].sample(10))
```
