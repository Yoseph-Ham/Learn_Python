# 범주형 변수

## 원-핫-인코딩 (가변수)

```python
import os
import pandas as pd
# 이 파일은 열 이름을 나타내는 헤더가 없으므로 header = None으로 지정하고
# "names" 매개변수로 열 이름을 제공합니다.
data = pd.read_csv('../data/adult.data', header = None, index_col=False,
                  names=['age', 'workclass', 'fnlwgt', 'education',  'education-num',
           'marital-status', 'occupation', 'relationship', 'race', 'gender',
           'capital-gain', 'capital-loss', 'hours-per-week', 'native-country',
           'income'])

data = data[['age', 'workclass', 'education', 'gender', 'hours-per-week',
            'occupation', 'income']]
display(data.head())
```

## 문자열로 된 범주형 테이터 확인하기

```python
print(data.gender.value_counts())
```

```python
print("원본 특성: \n", list(data.columns), "\n")
#pd.get_dummies()는 객체 타입이나 범주형을 가진 열을 자동으로 변환
data_dummies = pd.get_dummies(data)
print("get_dummies 후의 특성: \n", list(data_dummies.columns), '\n')
```

```python
data_dummies.head() # 더미로 변경한 후 데이터 확인
```

```python
data.info() # 오리지날 데이터
```

```python
# 모델을 학습시키기 전에 데이터로부터 타깃값(income)을 분리해야 한다.
features = data_dummies.loc[:, 'age':'occupation_ Transport-moving'] #income 부분 제외

X = features.values
y = data_dummies['income_ <=50K'].values
print("X.shape : {} y.shape : {}".format(X.shape, y.shape))

features.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
logreg = LogisticRegression()
logreg.fit(X_train, y_train)

print("테스트 점수 : {:.2f}".format(logreg.score(X_train, y_train)))
```

### 숫자로 표현된 범주형 특성

```python
# 숫자 특성과 범주형 문자열 특성을 가진 DataFrame을 만듭니다
import pandas as pd
demo_df = pd.DataFrame({'숫자 특성':[0, 1, 2, 1],
                       '범주형 특성':['양말', '여우', '양말','상자']})
display(demo_df)
```

```python
# get_dummies를 사용하면 문자열 특성만 인코딩되며 숫자 특성은 바뀌지 않는다.
display(pd.get_dummies(demo_df))
```

```python
# 숫자 특성도 가변수로 만들고 싶다면 컬럼 매개변수에 인코딩하고 싶은 열을 명시

demo_df['숫자 특성'] = demo_df['숫자 특성'].astype(str)
display(pd.get_dummies(demo_df, columns = ['숫자 특성', '범주형 특성']))
```

```python
# 4-2. OneHotEncoder 와 ColumnsTransformer : Sklearn으로 범주형 변수 다루기
# OneHotEncoder는 모든 열에 인코딩을 수행한다. (문자열, 정수 다 변환)
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse = False)
print(ohe.fit_transform(demo_df))
```

```python
#변환된 특성에 해당하는 원본 범주형 변수 이름 가져오기
print(ohe.get_feature_names())
```
