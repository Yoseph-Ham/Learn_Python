```python
import re # 정규식(Regular Express)

tmp_string = tmp_one.find(class_='sammyListing').get_text()

re.split(('\n|\r\n'), tmp_string) # split ==> 지정한 특정 패턴이 일치하면 분리 (| = or를 의미)

print(re.split(('\n|\r\n'), tmp_string)[0]) # 메뉴 이름
print(re.split(('\n|\r\n'), tmp_string)[1]) # 가게 이름
```

```python
from urllib.parse import urljoin

# 절대 경로로 잡힌 url은 그대로 두고 상대 경로로 잡힌 url은 절대 경로로 변경
```

```python
rank = []
main_menu = []
cafe_name = []
url_add = []

list_soup = soup.find_all('div', 'sammy')

for item in list_soup:
    rank.append(item.find(class_='sammyRank').get_text())
    
    tmp_string = item.find(class_='sammyListing').get_text()
    
    main_menu.append(re.split(('\n|\r|n'), tmp_string)[0])
    cafe_name.append(re.split(('\n|\r|n'), tmp_string)[1])
    
    url_add.append(urljoin(url_base, item.find('a')['href']))
```

### 주피터 노트북에서 상태 진행바를 쉽게 만들어주는 모듈
* pip install tqdm

```python
from tqdm import tqdm_notebook
import time

rank = [] # 순위
main_menu = [] # 메인 메뉴 이름
cafe_name = [] # 카페 이름
url_add = [] # 접근 주소

list_soup = soup.find_all('div', 'sammy')
bar_total = tqdm_notebook(list_soup)

for item in bar_total:
    rank.append(item.find(class_='sammyRank').get_text())
    
    tmp_string = item.find(class_='sammyListing').get_text()
    
    main_menu.append(re.split(('\n|\r|n'), tmp_string)[0])
    cafe_name.append(re.split(('\n|\r|n'), tmp_string)[1])
    
    url_add.append(urljoin(url_base, item.find('a')['href']))
    
    time.sleep(0.05)
```

```python
import pandas as pd

data = {'Rank':rank, 'Menu':main_menu, 'Cafe':cafe_name, 'URL':url_add} # 딕셔너리 생성
df = pd.DataFrame(data) # 데이터 프레임 생성
df.head()
```

# 맛집 위치를 지도에 표기하기

```python
import folium
import pandas as pd
import googlemaps
import numpy as np
```

```python
df = pd.read_csv('../data/03. best_sandwiches_list_chicago2.csv', index_col=0)
df.head(5)
```

```python
gmaps_key = 'AIzaSyCOc-RHpW3TSjrTheiRUDHh9e_xDkA3jfc'

gmaps = googlemaps.Client(key = gmaps_key)
```

```python
lat = []
lng = []

for n in tqdm_notebook(df.index):
    if df['Address'][n] != 'Multiple':
        target_name = df['Address'][n]+', '+'Chicago'
        gmaps_output = gmaps.geocode(target_name)
        location_output = gmaps_output[0].get('geometry')
        lat.append(location_output['location']['lat'])
        lng.append(location_output['location']['lng'])
    else:
        lat.append(np.nan)
        lng.append(np.nan)
        
df['lat'] = lat
df['lng'] = lng
df.head()
```

```python
mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()],
                    zoom_start = 11)
folium.Marker([df['lat'].mean(), df['lng'].mean()],
             pupup='center').add_to(mapping)

mapping
```

```python
mapping = folium.Map(location = [df['lat'].mean(), df['lng'].mean()], zoom_start = 11)


for n in df.index:
    if df['Address'][n] != 'Multiple':
        folium.Marker([df['lat'][n], df['lng'][n]],
                     popup = df['Cafe'][n]).add_to(mapping)
mapping
```

```python
mapping = folium.Map(location = [df['lat'].mean(), df['lng'].mean()], zoom_start = 11)

from folium.plugins import MarkerCluster
marker_cluster = MarkerCluster().add_to(mapping)

for n in df.index:
    if df['Address'][n] != 'Multiple':
        folium.Marker([df['lat'][n], df['lng'][n]],
                     popup = df['Cafe'][n]).add_to(marker_cluster)
mapping
```
