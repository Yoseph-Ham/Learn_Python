# 카카오 검색 API를 활용한 장소 검색

```python
# 카카오 검색 API 사용 예

import requests

searching = '합정 스타벅스'
url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(searching)
headers = {
    "Authorization": "KakaoAK 622abf8f99b2769c8370e5659effdc6b"       # 본인의 REST API키 입력
}

places = requests.get(url, headers = headers).json()['documents']
places
```

```python
def find_places(searching):
    url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(searching)
    headers = {
        "Authorization": "KakaoAK 622abf8f99b2769c8370e5659effdc6b"
    }

    places = requests.get(url, headers = headers).json()['documents']
    place = places[0]
    name = place['place_name']
    x = place['x']
    y = place['y']
    data = [name, x, y, searching]
    
    return data
```

```python
data = find_places('제주공항')
data
```

```python
from tqdm import tqdm_notebook
```

```python
import time
locations_inform = [ ]

for location in tqdm_notebook(locations):
    try:
        data = find_places(location)
        locations_inform.append(data)
        time.sleep(0.5)
    except:
        pass
    
locations_inform
```

```python
# 위치정보 저장
locations_inform_df = pd.DataFrame(locations_inform)
locations_inform_df.columns = ['name_official', '경도', '위도', '인스타위치명']
locations_inform_df.to_excel('../data/3_locations.xlsx', index = False)
```

```python
# 인스타 게시량 및 위치정보 데이터 불러오기
location_counts_df = pd.read_excel('../data/3_location_counts.xlsx', index_col = 0)
locations_inform_df = pd.read_excel('../data/3_locations.xlsx')
```

```python
# 위치 데이터 병합하기
location_data = pd.merge(locations_inform_df, location_counts_df,
                        how = 'inner', left_on = 'name_official', right_index = True)

location_data.head()
```

```python
# 데이터 중복 점검하기
location_data['name_official'].value_counts()
```

```python
# 장소 이름 기준 병합하기
location_data = location_data.pivot_table(index = ['name_official', '경도', '위도'],
                                         values = 'place', aggfunc = 'sum')

location_data.head()
```

```python
location_data.to_excel('../data/3_location_inform.xlsx')
```

## folium을 이용한 지도 시각화 -1 개별 표시

```python
location_data = pd.read_excel('../data/3_location_inform.xlsx')
location_data.info()
```

```python
import folium

Mt_Hanla = [33.362500, 126.533694]
map_jeju = folium.Map(location = Mt_Hanla, zoom_start = 11) # 한라산을 기준으로 줌

for i in range(len(location_data)):
    name = location_data ['name_official'][i] # 공식 명칭
    count = location_data['place'][i] # 게시글 개수
    size = int(count)*2
    long = float(location_data['위도'][i])
    lat = float(location_data['경도'][i])
    folium.CircleMarker((long,lat), radius = size, color = 'red', popup=name).add_to(map_jeju)

map_jeju
```
